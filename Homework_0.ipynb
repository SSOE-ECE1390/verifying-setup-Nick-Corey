```python
import cv2

import os
import numpy as np

import matplotlib.pyplot as plt
```

```python
# Load the image in using openCV 
apollo_image = cv2.imread(os.path.relpath('./Data/Apollo_11_Launch.jpg'), 0)

# View the Image as greyscale
plt.imshow(apollo_image, cmap="gray")
```

```python
# Crop the image
cropped_apollo = apollo_image[100:600, 200:1000]

plt.imshow(cropped_apollo, cmap="grey")
```

```python
# Brightening the image
matrix = np.ones(cropped_apollo.shape, dtype="uint8") * 100

bright_apollo = cv2.add(cropped_apollo, matrix)

plt.imshow(bright_apollo, cmap="gray")
```

```python
# Rotate the image 90 degrees
# wasn't in example notebook, found it here: https://www.geeksforgeeks.org/python/python-opencv-cv2-rotate-method/
rotated_apollo = cv2.rotate(bright_apollo, cv2.ROTATE_90_CLOCKWISE)

plt.imshow(rotated_apollo, cmap="grey")
```

```python
# Saving the Image
cv2.imwrite("./Data/Apollo_11_Launch_Modified.png", rotated_apollo)
```
